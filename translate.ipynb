```python
from googletrans import Translator

# Create an instance of the Translator class
translator = Translator()

# Enter the word you want to translate
word = input("Enter the word you want to translate: ")

# Use the translate() method to translate the word from Urdu to English
translation = translator.translate(word, src='en', dest='ur')

# Print the translated word
print(f"Translated word: {translation.text}")
```

```python

# Load the pipeline for translation
translator = pipeline("translation_en_to_ur", model="Helsinki-NLP/opus-mt-en-ur")

# Translate a sample sentence from English to Urdu
english_sentence = "Blue sky."
urdu_translation = translator(english_sentence, max_length=10)

# Print the Urdu translation
print(urdu_translation[0]['translation_text'])
```

```python
import os
from datasets import load_dataset

data_dir = "/content/umc005-corpus.zip" # replace with the path to the extracted files
file_names = ["test.en", "test.ur", "train.en", "train.ur"]
file_paths = [os.path.join(data_dir, file_name) for file_name in file_names]

# define the data format
data_format = {
    "train": {"en": [], "ur": []},
    "validation": {"en": [], "ur": []},
    "test": {"en": [], "ur": []}
}

# load the data from each file
for i, file_path in enumerate(file_paths):
    #encoding="utf-8"
    with open('/content/umc005-corpus.zip', "r") as f:
        lines = f.readlines()
        for line in lines:
            en, ur = line.strip().split("\t")
            data_format[file_names[i].split(".")[0]]["en"].append(en)
            data_format[file_names[i].split(".")[0]]["ur"].append(ur)

# create the dataset
dataset = load_dataset("custom", data_format=data_format)

```
